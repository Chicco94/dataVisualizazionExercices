```python
#%% Initials imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#%% Load the dataset and first exploratory analisys
iris = pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
print(iris.head())
```

```python
# MATPLOT LIB

# espongo i dati lunghezza del petalo rispetto alla larghezza

# distinguiamo i vari gruppi di fiori
colors = {'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'}

# creo una figura con relativi assi
fig, ax = plt.subplots()

# stampo punto a punto
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i],iris['sepal_width'][i],color=colors[iris['class'][i]])

# aggiungo relative etichette
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
```

```python
# PANDAS

# Assegno un colore in base alla classe
colors = {'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'}
iris['colors'] = [ colors[el] for el in iris['class']]

iris.plot.scatter(x='sepal_length',y='sepal_width', title='Iris Dataset', c='colors')
```
